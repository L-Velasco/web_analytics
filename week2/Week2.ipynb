```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
```

```python
# Add all nodes
G.add_nodes_from([2, 3])
```

```python
# Add edges
G.add_edge(1, 2)
G.add_edge(2, 3)
G.add_edge(4, 5)
G.add_edge(2, 5)
G.add_edge(2, 4)
G.add_edge(1, 4)
G.add_edge(2, 6)

labels = {1: 'Ben',
          2: 'Alexis',
          3: 'Carl',
          4: 'Marcus',
          5: 'Elizabeth',
          6: 'Andy'}
```

```python
# Print graph
nx.draw(G, labels=labels, node_size=1500,
                   node_color='lightgray')
```
